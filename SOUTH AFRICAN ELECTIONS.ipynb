```python
Gauteng = SA_elec_1999[SA_elec_1999.PROVINCE == 'GAUTENG']
```

```python
sns.set(style="ticks")
sns.pairplot(Gauteng, hue="PARTY_NAME")
```

```python
ANCvsDA = Gauteng[(Gauteng.PARTY_NAME == 'AFRICAN NATIONAL CONGRESS') | (Gauteng.PARTY_NAME == 'DEMOCRATIC PARTY')]
```

```python
ANCvsDA.head()
```

```python
plt.figure(figsize=(10, 7)) 
sns.lineplot(x="VALID_VOTES", y="TOTAL_VOTES_CAST",
             hue="PARTY_NAME", 
             data=ANCvsDA)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(Gauteng['VALID_VOTES'], Gauteng['PARTY_NAME'], cmap='rainbow')  
```

```python
Gauteng.head()
```

```python
# rename two of the columns by using the 'rename' method
Gauteng.rename(columns={'SPOILT VOTE':'SPOILT_VOTE'}, inplace=True)
Gauteng.columns
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
sns.set(style="whitegrid")
sns.scatterplot(x="REGISTERED_VOTERS", y="TOTAL_VOTES_CAST",
                palette="ch:r=-.2,d=.3_r",
                sizes=(1, 8), linewidth=0,
                data=Gauteng, ax=ax)
```
